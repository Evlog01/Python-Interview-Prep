## Задача 1.
Найти наименьшую степень 2, равную или больше суммы чисел в заданном массиве.

```python
def min_power_of_two(arr: list[int]) -> int:
    """
    Возвращает наименьшую степень двойки >= суммы массива.
    Временная сложность: O(N) на суммирование + O(1) на битовый сдвиг = O(N).
    """
    total = sum(arr)
    if total <= 1:
        return 1

    # Битовый сдвиг 1 << Y эквивалентен 2^Y.
    # (total - 1).bit_length() возвращает количество бит, необходимых для представления числа,
    # что математически эквивалентно поиску ближайшей степени двойки.
    return 1 << (total - 1).bit_length()

# === Тесты ===
test_arr = [1, 3, 5]
result = min_power_of_two(test_arr)
print(f"Сумма: {sum(test_arr)}, Наименьшая степень 2: {result}") # Ожидается: 16
```

* Временная сложность алгоритма O(N), зависит от количества элементов в массиве (N). Происходит вычисление суммы элементов.
* Временная сложность поиска степени O(1).

1. Проходим по всем элементам массива и складываем их
2. Если сумма <= 1, наименьшая степень двойки (2^0) равна 1
3. Для остальных случаев используем битовый сдвиг
4. Вычитаем 1, чтобы корректно обработать случаи, когда сумма уже является точной степенью двойки (например, 1, 2, 4, 8, 16)
5. Сдвиг единицы влево (1 << Y) эквивалентен 2 в степени Y

## Задача 2.
Дан массив целых чисел длины N и целое число K. Разрешена операция удаления первого элемента массива с записью в его конец K-го элемента. Какое минимальное число шагов потребуется для того, чтобы сделать все элементы массива одинаковыми? -1, если это невозможно.

```python
def min_steps_to_equal(arr: list[int], K: int) -> int:
    """
    Возвращает минимальное число шагов, чтобы все элементы стали одинаковыми.
    K задается в 1-индексации.
    Временная сложность: O(N).
    """
    n = len(arr)

    # 1. Первая проверка. Защита от некорректных данных
    if n == 0:
        return 0  # Пустой массив уже "одинаковый"
    if K < 1 or K > n:
        return -1 # K выходит за границы массива

    # 2. Определяем целевое значение
    # K-й элемент в 1-индексации имеет индекс K - 1 в Python
    target = arr[K - 1]

    # 3. Вторая проверка. Защита от цикла
    # Все элементы, начиная с K-го и до самого конца, УЖЕ должны быть равны target
    for i in range(K - 1, n):
        if arr[i] != target:
            return -1

    # 4. Подсчет шагов
    # Сдвигаем все элементы, которые стоят ПЕРЕД K-м элементом и не равны target
    # Идем справа налево, от элемента перед K-м (индекс K - 2) до начала массива (индекс 0)
    for i in range(K - 2, -1, -1):
        if arr[i] != target:
            # Если нашли такой элемент, количество шагов равно его индексу + 1
            return i + 1

    # 5. Если не нашли элемент в цикле выше, то все элементы перед K-м уже равны target
    return 0


# === Тесты ===
if __name__ == "__main__":
    print("Тестирование функции")

    # Обычные случаи
    print(f"[1, 2, 2, 2], K=2, {min_steps_to_equal([1, 2, 2, 2], 2)}")       # Ожидается: 1
    print(f"[3, 2, 2, 2], K=2, {min_steps_to_equal([3, 2, 2, 2], 2)}")       # Ожидается: 1
    print(f"[1, 2, 3, 4, 4], K=4, {min_steps_to_equal([1, 2, 3, 4, 4], 4)}") # Ожидается: 3

    # Бесконечный цикл
    print(f"[1, 2, 3, 2], K=2, {min_steps_to_equal([1, 2, 3, 2], 2)}")       # Ожидается: -1

    # Однородный массив
    print(f"[5, 5, 5], K=1, {min_steps_to_equal([5, 5, 5], 1)}")             # Ожидается: 0

    # Граничные условия
    print(f"[], K=1, {min_steps_to_equal([], 1)}")                           # Ожидается: 0
    print(f"[1, 2], K=5, {min_steps_to_equal([1, 2], 5)}")                   # Ожидается: -1 (K > N)
    print(f"[1, 2], K=0, {min_steps_to_equal([1, 2], 0)}")                   # Ожидается: -1 (K < 1)
```

## Задача 3.
Напишите функцию для решения следующей задачи:

В офисе AtCoder есть один увлажнитель. В момент времени 0 в нем нет воды.
Вы доливаете воду N раз. В i-й раз (1 ≤ i ≤ N) вы доливаете V_i литров в момент T_i.
Гарантируется, что T_i < T_{i+1} для всех 1 ≤ i ≤ N−1.
Увлажнитель протекает: пока в нем есть вода, её количество уменьшается со скоростью 1 литр за единицу времени.
Требуется найти количество воды в увлажнителе сразу после того, как вы закончите долив в момент T_N.

Данные подаются как многострочная строка (input_str):

Первая строка: одно целое число N — количество доливов.
Следующие N строк: по два целых числа T_i и V_i, разделённые пробелом.
Гарантируется, что T_i строго возрастают.
Функция должна вернуть строку с одним числом — количеством воды сразу после долива в момент T_N.

Ваш код должен пройти следующие тесты:

Тест 1
Вход:
4
1 3
3 1
4 4
7 1
Выход:
3

Тест 2
Вход:
3
1 8
10 11
21 5
Выход:
5

Тест 3
Вход:
10
2 1
22 10
26 17
29 2
45 20
47 32
72 12
75 1
81 31
97 7
Выход:
57

```python
def solve_water(input_str: str) -> str:
    """
    В момент времени 0 воды нет.
    В моменты t_i доливаем v_i литров.
    Вода утекает со скоростью 1 литр/единицу времени (но не уходит в минус).
    Возвращаем количество воды после последнего долива.

    Вычисления ведём дискретно от одного события долива к другому,
    находя, сколько воды утекло в промежутке.

    Временная сложность: O(N).
    """
    # Разбиваем входную строку на строки
    lines = input_str.strip().split('\n')
    n = int(lines[0])

    water = 0
    prev_time = 0

    # Проходим по всем N доливам
    for i in range(1, n + 1):
        t_i, v_i = map(int, lines[i].split())

        # 1. Вычисляем, сколько времени прошло с последнего события
        dt = t_i - prev_time

        # 2. Находим количество воды. Оно не может стать отрицательным
        water = max(0, water - dt)

        # 3. Добавляем новую воду
        water += v_i

        # 4. Обновляем время последнего события
        prev_time = t_i

    return str(water)


# === Тесты ===
if __name__ == "__main__":
    print("Тестирование функции")

    test1 = """4
1 3
3 1
4 4
7 1"""
    print(f"\nТест 1:\nВход:\n{test1}")
    print(f"Выход:\n{solve_water(test1)}")  # Ожидается: 3

    test2 = """3
1 8
10 11
21 5"""
    print(f"\nТест 2:\nВход:\n{test2}")
    print(f"Выход:\n{solve_water(test2)}")  # Ожидается: 5

    test3 = """10
2 1
22 10
26 17
29 2
45 20
47 32
72 12
75 1
81 31
97 7"""
    print(f"\nТест 3:\nВход:\n{test3}")
    print(f"Выход:\n{solve_water(test3)}")  # Ожидается: 57
```

## Задача 4.
Даны два строковых представления чисел A и B. Нужно максимизировать A, заменив в нём любую
цифру на цифру из B. Каждую цифру B можно использовать только один раз.

```python
def maximize_number(A: str, B: str) -> str:
    """
    Максимизируем число A, заменяя его цифры на цифры из B.
    Каждую цифру из B можно использовать только один раз.

    Временная сложность: O(N + M), где N – длина A, M – длина B.
    """
    # 1. Подсчитываем количество каждой цифры в B (от 0 до 9)
    # Это работает за O(M)
    counts = [0] * 10
    for char in B:
        counts[int(char)] += 1

    # 2. Проходим по цифрам числа A слева направо (от старших разрядов к младшим)
    # Чтобы максимизировать число, нам нужно заменить цифру в как можно более старшем разряде
    result = list(A)

    for i in range(len(result)):
        current_digit = int(result[i])

        # Ищем самую большую доступную цифру, которая строго больше текущей
        # Идем от 9 вниз до current_digit + 1
        for d in range(9, current_digit, -1):
            if counts[d] > 0:
                result[i] = str(d)
                counts[d] -= 1
                break # Цифра использована, переходим к следующей позиции в A

    return "".join(result)

# === Тесты ===
if __name__ == "__main__":
    print("Тестирование функции")
    # Обычные случаи
    print(f"A=123, B=4, result={maximize_number('123', '4')}")        # Ожидается: 423
    print(f"A=123, B=55, result={maximize_number('123', '55')}")      # Ожидается: 553 (заменяем обе)
    print(f"A=192, B=92, result={maximize_number('192', '92')}")      # Ожидается: 992 (1 на 9, 9 не меняем, 2 не меняем)
    print(f"A=111, B=999, result={maximize_number('111', '999')}")    # Ожидается: 999

    # Случаи, где замена не выгодна (все цифры в B меньше или равны)
    print(f"A=987, B=5, result={maximize_number('987', '5')}")        # Ожидается: 987
    print(f"A=456, B=123, result={maximize_number('456', '123')}")    # Ожидается: 456
```
